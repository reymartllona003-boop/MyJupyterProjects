Checking the distribution of math, writing and reading score

```python
sns.set(style="whitegrid")

# Create a figure with 3 plots side by side
plt.figure(figsize=(15, 5))

# Math Score
plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True, color=r'blue')
plt.title("Average Math Score")
plt.ylabel("Score")
plt.xlabel("Female/Male")

# Reading Score
plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True, color=r'red')
plt.title("Average Reading Score")
plt.ylabel("Score")
plt.xlabel("Female/Male")

# Writing Score
plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True, color=r'orange')
plt.title("Average Writing Scorer")
plt.ylabel("Score")
plt.xlabel("Female/Male")
```

```python
gender_avg = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()
gender_melted = gender_avg.melt(id_vars='gender',var_name='Subject',value_name='Average_Score')
plt.figure(figsize=[10,6])
sns.barplot(x='gender',y='Average_Score',hue='Subject', data=gender_melted)
plt.title("Average Score by Gender and Subject")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.legend(title="Subject")
plt.show()
```

Female students scored higher on average in reading and writing, while males performed slightly better in math.
