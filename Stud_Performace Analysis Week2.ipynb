STUDENT PERFORMACE 



Aim of the Analysis:

The aim of this analysis is to investigate the factors influencing student academic performance, particularly how parental education, gender, and test preparation impact grades in math, reading, and writing. The goal is to identify patterns and insights that can help improve student outcomes and guide educational support strategies.

Research Questions:
1. Does test preparation affect student performance?
2. Is there a gender difference in student performance?
3. How does parental education level influence student grades?

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df=df.rename(columns={'parental level of education': 'parental_education'})
```

Checking the distribution of math, writing and reading score

```python
sns.set(style="whitegrid")


plt.figure(figsize=(15, 5))


plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True, color=r'blue')
plt.title("Average Math Score")
plt.ylabel("Score")
plt.xlabel("Female/Male")


plt.subplot(1, 3, 2)
sns.histplot(df['reading score'], kde=True, color=r'red')
plt.title("Average Reading Score")
plt.ylabel("Score")
plt.xlabel("Female/Male")


plt.subplot(1, 3, 3)
sns.histplot(df['writing score'], kde=True, color=r'orange')
plt.title("Average Writing Scorer")
plt.ylabel("Score")
plt.xlabel("Female/Male")
```

The analysis shows that students who completed a test preparation course consistently scored higher in math, reading, and writing compared to those who did not. This indicates that exam preparation has a positive impact on student performance, helping them achieve better grades and improving overall academic outcomes.

```python
sns.set(style="whitegrid")

# Create a figure with 3 plots side by side
plt.figure(figsize=(15, 5))

# Math Score
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course', y = 'math score', data = df, palette='Blues')
plt.title("Math Scores vs Test Preparation")

# Reading Score
plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course', y = 'reading score', data = df, palette='Greens')
plt.title("Reading Scores vs Test Preparation")

# Writing Score
plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course', y = 'writing score', data = df, palette='Oranges')
plt.title("Writing Scores vs Test Preparation")
```

The number of male and female students who passed in math, writing, and reading scores was analyzed to compare their overall academic performance and identify patterns in subject strengths between genders. 

```python
df['Math_Pass'] = np.where(df['math score'] >= 75, 'Pass', 'Fail')


math_count = df[df['Math_Pass'] == 'Pass']['gender'].value_counts()

df['Reading_Pass'] = np.where(df['reading score'] >= 75, 'Pass', 'Fail')

reading_count = df[df['Reading_Pass'] == 'Pass']['gender'].value_counts()

df['Writing_Pass'] = np.where(df['writing score'] >= 75, 'Pass', 'Fail')


writing_count = df[df['Writing_Pass'] == 'Pass']['gender'].value_counts()

print(math_count)
print(reading_count)
print(writing_count)
```

The analysis shows that female students outperformed male students in most subjects. While 173 males passed in mathematics compared to 122 females, the trend reverses in the other subjects 240 females vs. 130 males in writing, and 241 females vs. 109 males in reading. This indicates that male students tend to perform better in math, whereas female students excel in language-related subjects such as writing and reading.

```python
gender_avg = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()
gender_melted = gender_avg.melt(id_vars='gender',var_name='Subject',value_name='Average_Score')
plt.figure(figsize=[10,6])
sns.barplot(x='gender',y='Average_Score',hue='Subject', data=gender_melted)
plt.title("Average Score by Gender and Subject")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.legend(title="Subject")
plt.show()
```

Analyzing the different status of educational level of students who passed the exam to understand their performance levels.

```python
df['Math_Pass_Int'] = df['Math_Pass'].map({'Pass': 1, 'Fail': 0})
Educ_lvl_Mathpass = df.groupby('parental_education')['Math_Pass_Int'].sum()

df['Reading_Pass_Int'] = df['Reading_Pass'].map({'Pass': 1, 'Fail': 0})
Educ_lvl_Readpass = df.groupby('parental_education')['Reading_Pass_Int'].sum()

df['Writing_Pass_Int'] = df['Writing_Pass'].map({'Pass': 1, 'Fail': 0})
Educ_lvl_Writepass = df.groupby('parental_education')['Writing_Pass_Int'].sum()

print(Educ_lvl_Mathpass)
print()
print(Educ_lvl_Readpass)
print()
print(Educ_lvl_Writepass)
```

Conclusion:

Based on the results, students whose parents attained an associates degree and some college education had the highest number of passes across all three subjects  math, reading, and writing. This suggests that moderate levels of parental education (some college or associates degree) may positively influence students academic performance.

Meanwhile, students whose parents held masters degrees showed comparatively fewer passes, which may indicate smaller sample size or other influencing factors such as workload or expectations. Overall, the data implies that parental education level plays a significant role in student achievement, particularly in math, reading, and writing performance.

```python
sns.set(style="whitegrid")

# Create a figure with 3 plots side by side
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
p = sns.countplot(x='parental_education', data = df, hue='Math_Pass', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
plt.title("Educational Level Math Assessment")

plt.subplot(1, 3, 2)
p = sns.countplot(x='parental_education', data = df, hue='Reading_Pass', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
plt.title("Educational Level Reading Assessment")

plt.subplot(1, 3, 3)
p = sns.countplot(x='parental_education', data = df, hue='Writing_Pass', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
plt.title("Educational Level Writing Assessment")
```

```python
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```

GRADE SUMMARY

```python
conditions = [
    (df['average_score'] > 90),
    (df['average_score'] > 80) & (df['average_score'] <= 90),
    (df['average_score'] > 70) & (df['average_score'] <= 80),
    (df['average_score'] > 60) & (df['average_score'] <= 70),
    (df['average_score'] > 50) & (df['average_score'] <= 60),
    (df['average_score'] <= 50)
]

grades = np.array(['A', 'B', 'C', 'D', 'E', 'F'], dtype=object)

df['Grade'] = np.select(conditions, grades, default='F')  # default is optional
```

```python
grade_summary = df.groupby('parental_education')['Grade'].value_counts().unstack(fill_value=0)

print(grade_summary)
```

Conclusion:

Higher parental education correlates with better student performance:

Students whose parents have a bachelors or masters degree tend to have more A and B grades and fewer D, E, and F grades.

For example, students with parents holding a masters degree have the highest proportion of A grades (6) relative to low grades (only 1 F).

Lower parental education is associated with more low grades:

Students whose parents have some high school education or only a high school diploma have more D, E, and F grades.

For instance, "some high school" parents show high counts of D (41), E (35), and F (31).

Middle categories show mixed performance:

Students with parents having associates degree or some college have moderate numbers of both high and low grades, suggesting other factors may also influence performance.

Overall Insight:
Parental education seems to have a significant influence on student grades. Higher parental education generally supports better academic performance, while lower parental education is linked to higher frequencies of lower grades.

```python
plt.figure(figsize=[10,6])

colors = ['#1f77b4',  # A - blue
          '#ff7f0e',  # B - orange
          '#2ca02c',  # C - green
          '#d62728',  # D - red
          '#9467bd',  # E - purple
          '#8c564b']  # F - brown

grade_summary.plot(kind='bar', color=colors)   
plt.title("Overall Grades by Educational Level")
plt.xlabel("Educational Level")
plt.ylabel("Number of Students")
plt.legend(title='Grades')
plt.show()
```
