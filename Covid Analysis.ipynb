```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Covid.csv')
df.head()   
```

```python
df.info
df.isnull().sum()
```

```python
df = df.rename(columns={
    'Country/Region': 'Country'
})
```

```python
top10 = df.nlargest(10, 'Confirmed')

plt.figure(figsize=(10,6))
plt.bar(top10['Country'], top10['Confirmed'], color='orange')
plt.title("Top 10 Countries with Most Confirmed COVID-19 Cases")
plt.xlabel("Country")
plt.ylabel("Number of Confirmed Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Death_to_Recovery_Ratio'] = df['Deaths'] / df['Recovered']
df[['Country', 'Deaths', 'Recovered', 'Death_to_Recovery_Ratio']].head()
```

```python
top10_ratio = df.sort_values('Death_to_Recovery_Ratio', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10_ratio['Country'], top10_ratio['Death_to_Recovery_Ratio'], color='red')
plt.title("Top 10 Countries by Death-to-Recovery Ratio")
plt.xlabel("Country")
plt.ylabel("Death/Recovery Ratio")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Confirmed'], df['Recovered'], color='blue')
plt.title("Confirmed vs. Recovered Cases (per Country)")
plt.xlabel("Confirmed Cases")
plt.ylabel("Recovered Cases")
plt.grid(True)
plt.show()
```

```python
top10_deaths = df.nlargest(10, 'Deaths')

plt.figure(figsize=(10,6))
plt.bar(top10_deaths['Country'], top10_deaths['Deaths'], color='purple')
plt.title("Top 10 Countries with Most Deaths")
plt.xlabel("Country")
plt.ylabel("Number of Deaths")
plt.xticks(rotation=45)
plt.show()
```
